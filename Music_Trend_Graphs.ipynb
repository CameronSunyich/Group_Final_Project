```python
# Import dependencies
import matplotlib
import matplotlib.pyplot as plt
import csv
import pandas as pd
import numpy as np
import scipy.stats as sts
```

```python
# Making sure plots are all inline and directly below code
%matplotlib inline
```

```python
# Load the data
data = "data.csv"
```

```python
# Read the data
data_df = pd.read_csv(data)
data_df.head()
```

```python
# Tempo Trend In Music

ax = plt.gca()

data_df.plot(kind='line',x='year',y='tempo',ax=ax)
plt.xlabel('Year')
plt.ylabel('Tempo')
plt.title('Tempo In Music Trend 1960-2019')
plt.legend()
plt.xlim(1960, 2019)
plt.show()
```

```python
# Duration of Music Overtime (1960-2019)

ax = plt.gca()

data_df.plot(kind='line',x='year',y='duration_ms',ax=ax)
plt.xlabel('Year')
plt.ylabel('Duration (Minutes)')
plt.title('Duration of Music 1960-2019')
plt.legend()
plt.xlim(1960, 2019)
plt.show()
```

```python
# "Loudness" of Music Trend 1960-2019

ax = plt.gca()

data_df.plot(kind='line',x='year',y='loudness',ax=ax)
plt.xlabel('Year')
plt.ylabel('Loudness')
plt.title('"Loudness" Trend of Music 1960-2019')
plt.legend()
plt.xlim(1960, 2019)
plt.show()
```

```python
# Dancebility of Music 1960-2019

ax = plt.gca()

data_df.plot(kind='line',x='year',y='danceability',ax=ax)
plt.xlabel('Year')
plt.ylabel('Danceability')
plt.title('Danceability of Music 1960-2019')
plt.legend()
plt.xlim(1960, 2019)
plt.show()
```

```python
# Energy Trend of Music 1960-2019

ax = plt.gca()

data_df.plot(kind='line',x='year',y='energy',ax=ax)
plt.xlabel('Year')
plt.ylabel('Energy')
plt.title('Energy Trend of Music 1960-2019')
plt.legend()
plt.xlim(1960, 2019)
plt.show()
```

```python
# Explicitness of Music 1960-2019

ax = plt.gca()

data_df.plot(kind='line',x='year',y='explicit' ,marker="*" ,linewidth=2 ,ax=ax)
plt.xlabel('Year')
plt.ylabel('Explicitness')
plt.title('Explicitness of Music 1960-2019')
plt.legend()
plt.xlim(1960, 2019)
plt.show()
```
