```python
import pandas as pd
import numpy as np

from sqlalchemy import create_engine
import psycopg2

from config import db_password
```

```python
music_df = pd.read_csv('cleaned_music_data.csv')
music_df.rename(columns={'Unnamed: 0':'orig_index'},inplace=True)
music_df = music_df.set_index('orig_index')
music_df.head()
```

```python
# Create the connection to the PostgreSQL database
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/music_data"
engine = create_engine(db_string)

# Add data to a SQL database 
music_df.to_sql(name='music_data', con=engine, if_exists='replace')
```
