```python
import os
import pandas as pd

path = os.path.join('C:/Users/aweso/Desktop/UO-Bootcamp/Final_Project/Resource/data.csv')
music_data = pd.read_csv(path)
music_data
```

```python
cleaned_music_data = music_data[music_data['year'] >= 1960]
cleaned_music_data = cleaned_music_data[cleaned_music_data['year'] <= 2019]
#drop mode, id, speechiness, popularity?, liveness, valence, song released?, key?
cleaned_music_data = cleaned_music_data.drop(['id'],axis = 1)
cleaned_music_data = cleaned_music_data.drop(['mode'],axis = 1)
cleaned_music_data = cleaned_music_data.drop(['speechiness'],axis = 1)
cleaned_music_data = cleaned_music_data.drop(['liveness'],axis = 1)
cleaned_music_data = cleaned_music_data.drop(['valence'],axis = 1)
cleaned_music_data = cleaned_music_data.drop(['key'],axis = 1)
cleaned_music_data = cleaned_music_data.drop(['name'],axis = 1)
cleaned_music_data = cleaned_music_data.drop(['artists'],axis = 1)
cleaned_music_data
```

```python
cleaned_music_data.to_csv('cleaned_music_data.csv', sep=',')
```
